```python
# Passo 1: Importar a base de dados
# Passo 2: Abrir ler e Entender a base de dados (encontrar problemas na base de dados)
# passo 3: Resolver os problemas da base de dados (limpeza, tratamento)
    # Problemas da base
        # - Dados faltando
        # - Dados duplicados
        # - Dados incorretos
        # - Formatação
        # - Dados Vazios
        # - Outliers              
# Passo 4: Analisar a base de dados (encontrar padrões, tendências, etc, caso de cancelamentos de clientes e %)
# Passo 5: Análise da causa dos cancelamentos (por que os clientes estão cancelando)



# Download Pandas, plotly, 

# Passo 1: Importar a base de dados
import pandas as pd

# Passo 2: Abrir ler e Entender a base de dados (encontrar problemas na base de dados)
tabela = pd.read_csv("cancelamentos.csv")

display(tabela.info())  # Verificar informações gerais da tabela

tabela = tabela.drop(columns = 'CustomerID')    # Remover coluna que não agrega valor
# --------------------------------------------------------------------------------------
# tabela.fillna("Não informado", inplace=True)  # Preencher valores vazios com "Não informado"
# tabela.ffill(inplace=True)  # Preencher valores vazios com o valor da célula acima
#---------------------------------------------------------------------------------------
tabela = tabela.dropna()  # Remover linhas com conteúdo vazio

display(tabela.info())
display(tabela)


# Passo 4: Analisar a base de dados (encontrar padrões, tendências, etc, caso de cancelamentos de clientes e %)

display(tabela["cancelou"].value_counts())  # Contar quantos clientes cancelaram e quantos não cancelaram

display(tabela["cancelou"].value_counts(normalize =True)) # Contar % de clientes que cancelaram e que não cancelaram

#display(tabela["cancelou"].value_counts(normalize =True).map("{:.2%}".format))  # Contar % de clientes que cancelaram e que não cancelaram com numeros formatados

# Passo 5: Análise da causa dos cancelamentos (por que os clientes estão cancelando)
    # Entender as causas através do cruzamento de dados
    # criar gráficos para facilitar a visualização dos dados
        # Passo 1 criar o gráfico
    
import plotly.express as px 

# grafico = px.histogram(tabela, x="idade", color="cancelou")

for coluna in tabela.columns:
    if coluna != "cancelou":
       print(f"Analisando a coluna {coluna}")
    else:
       break
    grafico = px.histogram(tabela, x=coluna, color="cancelou", text_auto=True)    
    
    # Passo 2: analisar o gráfico e tirar conclusões

    grafico.show()

# analisar problemas de cancelamentos com relação ao call center
tabela = tabela[tabela["ligacoes_callcenter"]<5]    
    
# analisar problemas de cancelamentos com relação ao tempo de contrato
tabela = tabela[tabela["duracao_contrato"]!="Monthly"]    

# analisar problemas de cancelamentos com relação ao atraso no pagamento
tabela = tabela[tabela["dias_atraso"]<21]

# visualizar a diferença entre os cancelamentos após a tratativa de erros da empresa
display(tabela["cancelou"].value_counts(normalize=True).map("{:.2%}".format))
```
